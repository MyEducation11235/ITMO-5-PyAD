\Кстати, давайте **переименуем** для удобства столбцы на русский

```python
# в этом поможет метод rename,
# не забываем сделать пересохранение!
# тут есть inplace=True !
data = data.rename(columns={
    'PassengerId': 'ID',
    'Survived': 'Выживаемость',
    'Pclass': 'Класс',
    'Name': 'Имя',
    'Sex': 'Пол',
    'Age': 'Возраст',
    'SibSp': 'Число_родств_1',
    'Parch': 'Число_родств_2',
    'Ticket': 'Билет',
    'Fare': 'Тариф',
    'Cabin': 'Каюта',
    'Embarked': 'Порт'
})
data.head()
```

Неинформативные или ненужные столбцы можно **удалять**:

```python
# в этом поможет метод drop
# главное - указывайте axis=1, чтобы удалить столбец
# (axis=0) отвечает за удаление строк
data = data.drop('ID', axis=1)
# data.drop('ID', axis=1, inplace=True) # если не хотите пересоздавать копию!
data.head()
```

А можно и **создавать новые**:

```python
data['Суммарное_число_родств'] = data['Число_родств_1'] + data['Число_родств_2']
```

Ещё иногда в данных встречаются **дубликаты**, что может мешать исследованиям:

```python
# метод drop_duplicates убирает дубликаты из данных
# тут есть inplace=True !
print(data.shape)
data = data.drop_duplicates()
print(data.shape)
```

Часто бывает необходимо **отсортировать** таблицу по какому-либо столбцу/столбцам.

В параметрах метода можно уточнить эти столбцы, а также порядок сортировки - по возрастанию или убыванию (*ascening: True/False*):


```python
# сортировка по ОДНОЙ колонке ВОЗРАСТА в порядке УБЫВАНИЯ
data.sort_values('Возраст', ascending=False).head(10)
```

```python
# сортировка по ДВУМ колонкам КЛАССА и ВОЗРАСТА
# в порядке УБЫВАНИЯ и ВОЗРАСТАНИЯ соответственно
data.sort_values(['Класс', 'Возраст'], ascending=[False, True]).head(10)
```

Также нередко приходится выполнять **фильтрацию**, здесь главное -- правильно записать условие. Принцип почти тот же самый, что и в обычных `if`:

* каждое *атомарное условие* возвращает `True` или `False`;
* условие может быть *комплексным* -- т.е. состоять из нескольких *атомарных условий*, которые соединяются через "и", "или"; здесь важное отличие от `if`: **"и" записывается как `&`, "или"  как `|`**;
* чтобы избежать ошибок и неточностей, лучше обрамлять *атомарные условия* в отдельные скобки.

```python
# можно отфильтровать по любому условию
data[(data['Класс'] == 3) | (data['Пол'].isin(data['Имя'].values[:10]))].shape
```

```python
data['Имя'].values[:10]
```

```python
# пример полезного применения:
# посчитаем средний возраст женщин и мужчин
data_male = data[data['Пол'] == 'male']
data_female = data[data['Пол'] == 'female']
print('Средний возраст женщин:', data_female['Возраст'].mean())
print('Средний возраст мужчин:', data_male['Возраст'].mean())
```

При помощи Pandas любые данные можно также **отрисовать** -- это поможет нам в выявлении закономерностей. Хоть Pandas и не предназначена для сложных  визуализаций, многие базовые возможности в ней имеются! Пока без погружения в сложные математические детали, держите несколько популярных графиков (указываем разные `kind` в методе `plot`):
* **Линия** (`kind='line'` + параметры `x`, `y`). Обратите внимание, что линия будет построена в точности по тем данным, которые Вы предоставили -- т.е. с точностью до порядка. Поэтому чтобы не получить траектории броуновского движенияя частиц при высокой температуре -- лучше предварительно отсортировать датафрейм:)

[*ссылка на документацию метода*](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html), тут можно посмотреть возможные параметры



```python
data.sort_values('Возраст').plot(x='Возраст', y='Тариф', kind='line', title='Цена билета в зависимости от возраста')
```

* **Столбчатая диаграмма** (`kind='bar'`). Логичная особенность этого метода в том, что данные нужно предварительно сгруппировать, выбрав подходящую агрегирующую функцию.

[*ссылка на документацию метода*](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html)

```python
data.groupby('Класс')['Возраст'].mean().plot(kind='bar', title='Средний возраст пассажиров по классам', color='r', legend=True)
```

* **Гистограмма** (или график распределения, `kind='hist'`). График распределения -- это "одномерная" зависимость (т.е. которая строится из одного признака), поэтому тут сразу применяем метод только к одной выбранной колонке (не ко всему датафрейму).

[*ссылка на документацию метода*](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.hist.html)

```python
data['Возраст'].plot(kind='hist', bins=50, title='Распределение возраста пассажиров')
```

* **Диаграмма рассеяния** (`kind='scatter'`). Можно ничего не сортировать, т.к. все наблюдения отобразятся просто точками. Понадобятся две оси: `x` и `y`.

[*ссылка на документацию метода*](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.scatter.html)

```python
data.plot(kind='scatter', x='Возраст', y='Тариф', alpha=0.5, title='Диаграмма рассеяния: цена билета vs возраст')
```

Также можно **объединять** несколько датафреймов в один -- это часто происходит на этапе сбора полезной информации, которая может содержаться в разных таблицах, а также во время *feature engineering*.


```python
# cоздаем второй датафрейм - со средней выживаемостью по классам
survival_rates = data.groupby('Класс')['Выживаемость'].mean().reset_index()
survival_rates = survival_rates.rename(columns={'Выживаемость': 'Доля_выживаемости'})
print(survival_rates.shape)
survival_rates.head()
```

```python
# cоздаем второй датафрейм - со средней выживаемостью по классам
survival_rates = data.groupby('Класс')['Выживаемость'].mean().reset_index()
survival_rates = survival_rates.rename(columns={'Выживаемость': 'Доля_выживаемости'})

# объединяем два DataFrame по классу пассажира
merged_df = data.merge(survival_rates, on='Класс', how='left')
merged_df.head()
```

здесь показан пример, когда столбцы, по которым мы делаем объединение, называются в обеих таблицах одинаково. Если это не так, можно вместо единого параметра `on` использовать отдельные `left_on` и `right_on`.
